```python
# Define positive and negative words
positive_words = ['happy', 'good', 'awesome', 'excellent', 'love', 'amazing', 'beautiful', 'brilliant', 'cheerful', 'delightful', 'fantastic', 'friendly', 'great', 'impressive', 'joyful', 'kind', 'lovely', 'marvelous', 'nice', 'outstanding', 'perfect', 'pleasant', 'positive', 'remarkable', 'spectacular', 'superb', 'wonderful']
negative_words = ['sad', 'bad', 'terrible', 'hate', 'awful', 'boring', 'cruel', 'depressing', 'disappointing', 'disgusting', 'dreadful', 'evil', 'faulty', 'frustrating', 'gross', 'horrible', 'hurtful', 'lousy', 'messy', 'nasty', 'negative', 'poor', 'rude', 'ugly', 'worst']
```

```python
def analyze_sentiment(text):
    # Convert input to lowercase for case-insensitive matching
    text = text.lower()

    # Split the text into words
    words = text.split()

    # Check if any positive or negative word is present
    positive_found = any(word in positive_words for word in words)
    negative_found = any(word in negative_words for word in words)

    if positive_found and not negative_found:
        return "Positive Sentiment"
    elif negative_found and not positive_found:
        return "Negative Sentiment"
    elif positive_found and negative_found:
        return "Mixed Sentiment"
    else:
        return "Neutral Sentiment"
```

```python
# Take input from user
user_input = input("Enter a sentence to analyze sentiment: ")

# Analyze and print the sentiment
result = analyze_sentiment(user_input)
print("Sentiment:", result)
```
